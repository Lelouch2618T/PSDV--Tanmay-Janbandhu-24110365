```python
l = []                                        # 1. a)
for j in range(1,21):
    l.append(j)
d = 0
e = 0
for i in l:
    if i == 1:
        e+=i
    elif i%2==0:
        d+=i
    else:
        e+=i
print(f'{d} is the sum of even numbers')      # 1. b)
print(f'{e} is the sum of odd numbers')       # 1. c)
mean = sum(l)/20                              # 1. d)
median = (l[len(l)//2] + l[(len(l)//2)+1])/2               
x = 0
for k in l:
    x += (k-mean)**2
stddev = (x/19)**0.5
print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'Standard Deviation: {stddev}')
```

```python
import matplotlib.pyplot as plt
def fibonacci(x):
    i = 0
    k = 1
    l = [0, 1]
    for j in range(1, x-1):
        s = i + k
        i = k
        k = s
        l.append(s)
    return l
n = int(input('Enter a number: '))
print(fibonacci(n))
plt.plot(fibonacci(n), linewidth = 2)
plt.show()
```

```python
import matplotlib.pyplot as plt
def countchar(str2):
    string_ = str2.lower().replace(' ', '')
    d = {}
    for i in string_:
        cnt = string_.count(i)
        d[i] = cnt
    return d
strin = input()
d = countchar(strin)
plt.bar(d.keys(), d.values())
plt.show()
```
